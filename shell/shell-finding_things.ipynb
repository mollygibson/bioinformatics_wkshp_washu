## Creating files with the `nano` text editor
In this workshop, we'll be writing several Python scripts.  You're probably most familiar with writing documents in Microsoft Word/Pages/Google Docs/etc.  When we write a script, it's best to write it in a specialized program called a **text editor**. For this workshop, we're going to use the text editor called **nano**.  While nano is pretty basic, it has the advantage that it's very easy to pick up.  When you become a more advanced coder, we recommend [Sublime Text](http://www.sublimetext.com/) for Mac and Windows users.

## Getting remote files
So far, all the files/directories we've created ourselves (with nano) or were already on the file system (e.g., our home directories).  What if the file we want is not on the remote computer?  There are two main ways to get files:

* Downloading files from your local computer with `Cyberduck`
* Downloading files from the Internet with `wget`

To look at the contents of the file, we could use the `cat` command like we learned previously.  However, one of the downsides of using `cat` is that it prints the entire file.  This may be okay for small files, but it can become a overwhelming for large files.  Two commands that are better suited for seeing small chucks of files are **`head`** and **`tail`**.

`head` prints the first 10 lines of a file.  The syntax is:

        $ head file
        
where `file` is the file you want to view the contents of.

Can you guess what `tail` does?  `tail` prints the last 10 lines of a file.  The syntax is:

        $ tail file
        
where `file` is the file you want to view the contexts of.

## Finding things
In this section, we'll learn how to find text within files and how to find specific files and directories in the file system.

### Finding text within files
To find specific text in a file, use the **`grep`** command.  The syntax for `grep` is:

        $ grep [options] "pattern_to_search_for" filename_to_search
        
where `[options]` are optional options that specify how to perform the search, `pattern_to_search_for` is the text you want to search for in quotes and `filename_to_search` is the name of the file that you want to find the text in.  Notice that `options` was enclosed in `[]`, this means that this part of the command is optional.

Let's find the average height in the United states:

        $ grep "United States of America" heights_by_country.tsv
        
`grep` has a lot of useful options; here are some of them:

Option | Description
--- | ---
`-i` | Performs a case-INsensitive search
`-v` | Print the lines that do NOT match the pattern
`--color`| Color the matching text
`-n` | Prefix each line of output with the line number within its input file

**TIP:** use the **`man`** command to learn more about a command.  The syntax for `man` is:

        $ man command_name
        
where `command_name` is the command you want to learn about.  `man` stands for *<span style="color:red">man</span>ual*.  Use the up and down arrow keys to scroll through the reference manual.  Type <kbd style="color:#000;background:#eee;border-style:solid;border-color:#ccc #aaa #888 #bbb;padding:.1em .6em;border:1px solid #ccc;font-size:11px;font-family:'Helvetica Neue',Helvetica,Arial,sans-serif;background-color:#f7f7f7;color:#333;-moz-box-shadow:0 1px 0 rgba(0,0,0,0.2),0 0 0 2px #fff inset;-webkit-box-shadow:0 1px 0 rgba(0,0,0,0.2),0 0 0 2px #fff inset;box-shadow:0 1px 0 rgba(0,0,0,0.2),0 0 0 2px #fff inset;border-radius:3px;display:inline-block;margin:0 .1em;text-shadow:0 1px 0 #fff;line-height:1.4;white-space:nowrap">q</kbd> to <span style="color:red">q</span>uit out of the reference manual.

### Finding files and directories
To find specific files and directories, use the **`find`** command.  The syntax for `find` is:

        $ find [path_to_search] [criteria_for_searching] [what_to_do]

where `path_to_search` is the path you want to search in.  By default, it is the current working directory.  `criteria_for_searching` if how you want to filter the search results, e.g., filter for files that were modified today, etc.  By default, no filtering is done.  `what_to_do` is what you want to do with the search results, e.g., delete them (yikes!).  By default, the search results are printed.

Before we get started, change to your home directory:

        $ cd ~

To start off simple, we're going to find where we saved the file called `english.txt`:

        $ find -name "english.txt"
        ./shell_tutorial/data/english.txt
        
This `find` command searches recursively in the current working working directory for a thing called `english.txt` if it finds anything, it prints the filename.  In the output, you can see that the command found our file.  On close inspection, you can see a `./` at the beginning of the path.  The `./` is shorthand for the current working directory.  So, our file is located in a the `data` directory which is in the `shell_tutorial` directory which is in the current working directory.

Next, let's find all `.txt` files.  We can do this with another powerful Unix feature called **wildcards**.  A wildcard is a character that stands for a class of characters in pattern matching.  The **`*` wildcard** matches zero or more characters.  This will hopefully make more sense with an example!

        $ find -name "*.txt"
        ./my_haiku.txt
        ./shell_tutorial/data/spanish.txt
        ./shell_tutorial/data/english.txt
        ./shell_tutorial/data/cat.txt

This `find` command searches recursively in the current working working directory for anything that ends in `.txt`.

Finally, let's find all files that were modified in the last 24 hours:

        $ find -type f -mtime -1
        ./my_haiku.txt

The `-type f` option filters the results for just files (as opposed to files and directories) and the `-mtime -1` filters for files that were modified within in the last 1*24 hours.

### Putting it together: finding text within files in `find` search results
In this section, we're going to combine the previous two sections to find specific text within our `find` searcg results.  Hold on your hats, it's about to get a little more complex!

Let's find all words that are contained in `.txt` files that contain the letter o:

        $ find -type f -name "*.txt" -exec grep "o" {} \;
        Bonjour
        Hello
        Meow

The first part of the command we're familiar with: it searches recursively in the current working working directory for any file that ends in `.txt`.  The latter part of the command `-exec grep "o" {} \;` means execute the command `grep "o"` on each search result and return the output from that command.  The `{}` is used as a placeholder for the search result and the `\;` denotes that the end of the command.
