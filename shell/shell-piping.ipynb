# Redirecting and piping

## Standard I/O streams
In a Unix shell, a **stream** is a sequence of characters that can be read or written by a command.  There are three standard input/output streams:

Stream | **File descriptor** | Description
---|:---:|---
**Standard input stream** (**stdin**) | 0| The input for a command
**Standard output stream** (**stdout**)| 1| The output from a command.
**Standard error stream** (**stderr**)| 2| The error output from a command.

We'll explain what a file descriptor is in the next section.

## Redirecting output
By default, the stdin and stdout streams are printed to the terminal window.  When there is a lot of text that is printed and/or we want to save what is printed, we can override this behavior with **stream redirection** and print the streams to files. 

There are two methods for redirecting a stream to a file:
* Overwriting with **`file_descrptor> file`**.
* Appending with **`file_descrptor>> file`**.

Before we get started with executing commands, change to the `~/shell_tutorial directory`:

        $ cd ~/shell_tutorial
        
Try this redirection command:

        $ ls my_haiku.txt XXX.txt 1>ls.out 2>ls.error

The command above checks if two files (`my_haiku.txt` and `XXX.txt`) exist in the current working directory using `ls`.  Any output is printed to the file `ls.out` and any error is printed to the file `ls.error`.  Since `my_haiku.txt` exists in the current working directory, the filename should be printed to `ls.out`.  Since `XXX.txt` does not (yet) exist, `ls` should print an error stating that the file doesn't exist.  Let's check that this is correct:

        $ cat ls.out
    my_haiku.txt
    $ cat ls.error
        ls: cannot access XXX: No such file or directory

If you want to send both the stdin and the stdout to the same file, use the special file descriptor **`&`**:

        $ ls my_haiku.txt XXX.txt &>output_and_error
    $ cat output_and_error
        ls: cannot access XXX.txt: No such file or directory
        my_haiku.txt

# Joining commands together with pipes
**Pipes** are used to redirect the stdout from one command to the stdin of the next command.  This creates a **pipeline**.  Pipelines are useful to create complex commands using simple commands.  Pipes are created with the `|` character (<kbd style="color:#000;background:#eee;border-style:solid;border-color:#ccc #aaa #888 #bbb;padding:.1em .6em;border:1px solid #ccc;font-size:11px;font-family:'Helvetica Neue',Helvetica,Arial,sans-serif;background-color:#f7f7f7;color:#333;-moz-box-shadow:0 1px 0 rgba(0,0,0,0.2),0 0 0 2px #fff inset;-webkit-box-shadow:0 1px 0 rgba(0,0,0,0.2),0 0 0 2px #fff inset;box-shadow:0 1px 0 rgba(0,0,0,0.2),0 0 0 2px #fff inset;border-radius:3px;display:inline-block;margin:0 .1em;text-shadow:0 1px 0 #fff;line-height:1.4;white-space:nowrap">shift</kbd>+<kbd style="color:#000;background:#eee;border-style:solid;border-color:#ccc #aaa #888 #bbb;padding:.1em .6em;border:1px solid #ccc;font-size:11px;font-family:'Helvetica Neue',Helvetica,Arial,sans-serif;background-color:#f7f7f7;color:#333;-moz-box-shadow:0 1px 0 rgba(0,0,0,0.2),0 0 0 2px #fff inset;-webkit-box-shadow:0 1px 0 rgba(0,0,0,0.2),0 0 0 2px #fff inset;box-shadow:0 1px 0 rgba(0,0,0,0.2),0 0 0 2px #fff inset;border-radius:3px;display:inline-block;margin:0 .1em;text-shadow:0 1px 0 #fff;line-height:1.4;white-space:nowrap">\</kbd>).  The syntax for piping commands is:

        $ command1 | command2 | command3 ...

You can string as many commands as you want together.

Let's practice piping with the `heights_by_country.tsv` file.  First change to the `~/shell_tutorial` directory

        $ cd ~/shell_tutorial
        
Then execute this command:

        $ sort -k2n heights_by_country.tsv | head
        Papua_New_Guinea        152.4   160.9
        Viet_Nam        157.7   160
        Japan   157.8   171.7
        Indonesia       158.6   164
        Indonesia_(until_1999)  158.6   164
        Philippines     158.9   165.6
        Guyana  159.2   170.5
        Democratic_People's_Republic_of_Korea   160.1   173.8
        Republic_of_Korea       160.1   173.8
        Myanmar 160.7   161.3

What's happening behind the scenes: the command **`sort`** sorts the file `heights_by_country.tsv` by the second column numerically (the 1880 height) and outputs the sorted file to the stdout stream.  This output is then fed into the `head` command which prints the first 10 lines of the stdin to stdout.  Since this stream is not piped to another command or to a file, it's printed to the terminal window.

# Getting yourself out of trouble
As we continue to learn more complex commands, it's important to have a few safety nets is place in case we get into trouble.  For example, when we were learning how to navigate the file system, we learned that that `cd ~` will always take us to our nice, cozy home.  Here are a few more safety nets to make you feel nice and cozy:
