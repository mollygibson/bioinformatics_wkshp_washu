# Pipelines: Putting it all together

In this final lesson, we're going to put a lot of what we've learned together, literally, to create a **pipeline**.  

## What is a pipeline?
A pipeline is a sequence of analysis steps where the output of one step is used as the input for the next step.  The word 'pipeline' comes from an analogy with plumbing: in a pipe, fluid travels in only one direction.  In an analysis pipeline, the steps are executed in one direction, namely, from the beginning of the pipeline to the end.  You've probably come across pipelines in your everyday life, but just didn't call them that.  For example, you may have used a *getting ready for the workshop* pipeline this morning:

<img src="pipeline1.png" height="100">

In this pipeline, we first wake up, then eat breakfast, then go to the workshop.  The steps are followed in a specific, logical order.

In bioinformatics, we often have to perform the same set of analysis steps on different datasets.  The tune sort of goes like this: get the data, preprocess the data, process the data, and visualize the data.  If we were to manually complete each one of these steps for each of the datasets it would be 1) very tedious 2) error-prone.  Instead, if we had a script to run all of these steps together it would be 1) very easy to analyze the data and 2) less error-prone.  Scripts that perform the steps of a pipeline are called **wrapper scripts** because they "wrap" all the steps together into one script.  Analyzing data with wrapper scripts makes the analysis reproducible.  This is because running the script multiple times on the same input should always produce the same output.

## Writing your first wrapper script
We're going to write a wrapper script in Python to analyze cd-hit data.  The pipeline will have the following steps:

1. Run `cd-hit` on a protein sequences file.
2. Run a `cd-hit` tool called `clstr_size_stat.pl` to calculate the distribution of cluster sizes
3. Run a Python script to plot the distribution of cluster sizes

Thus, the input to our pipeline will be a file containing protein sequences and the output will be a figure in pdf format.  Graphically, it looks like this:

<img src="pipeline2.png" height="700">

We're going to shove all of these steps into a single script called `cd_hit_wrapper.py`.

Before we start writing any code, it's time for a little philosophy about writing code.  When writing code, there are two main approaches: [top-down and bottom-up](http://en.wikipedia.org/wiki/Top-down_and_bottom-up_design).  In top-down programming, you first outline the major steps you need to do, then you successively divide the steps into smaller and smaller steps until you are left with the simplest of steps that you can easily code.  Bottom-up programming is the opposite: you code all of the basic steps, and then combine them together to make the more complex steps.  We're going to use the top-down approach because it has the advantage that we'll be able to see the big picture of our program early on. This means that during the development process, we may be calling scripts that do not yet exist.  By the end of the process, we will have coded all the steps so the program will be functional.

**Hints:**
* Use the Python function **`subprocess.call`** to call an external command from within a Python script.  The syntax is:

        subprocess.call(list_of_arguments)
    
  where `list_of_arguments` is a Python list of the words in the command.
    For example, if you wanted to execute
    
        $ cd-hit -i proteins.faa -o clustered_proteins.faa
        
    `list_of_arguments` would be
    
        ["cd-hit", "-i", "proteins.faa", "-o", "clustered_proteins.faa"]
        
    and the `subprocess.call` code would be
    
        subprocess.call(["cd-hit", "-i", "proteins.faa", "-o", "clustered_proteins.faa"])
        
* When you want to capture the stdout of an external command, the syntax for `subprocess.call` is a little different:

        subprocess.call(list_of_arguments, stdout=file_object)

  where `file_object` is a Python file object for a file that you opened for writing.  Refer to the previous [reading and writing files lesson](http://nbviewer.ipython.org/github/mollygibson/bioinformatics_wkshp_washu/blob/master/python/lessons/Python7_FileIO.ipynb) for help with file objects.
