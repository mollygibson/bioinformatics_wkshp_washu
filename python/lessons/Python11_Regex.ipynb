# Regular expressions

Before we dive into using regular expressions in Python, we first need to learn what regular expressions are.

## What are regular expression?

A **regular expression** is a string formatted using a special pattern notation to match another string.  Regular expressions are a very powerful tool for parsing text.  

You've actually seen a flavor of regular expressions before when we learned about Unix wildcards.  This command

        $ ls *.txt
        
finds all filenames that end in `.txt`.  In this example, we're checking if a string (a filename) matches a pattern (ends in `.txt`).


A quick heads up: the notation for regular expressions can become quite ugly.  Jamie Zawinski once famously said:

> Some people, when confronted with a problem, think "I know, I'll use regular expressions." Now they have two problems.

The regular expression notation can get ugly because we're trying write the patterns as strings using just the symbols on a keyboard instead of say, inventing new symbols the way mathematicians often do.  Most of the characters in the notation are treated literally, e.g., an a will match an a, a 1 will match a 1, etc.  There are a few handfuls of special characters called **metacharacters** that don't match their literal character.  These metacharacters are what allows regular expression to be so powerful.  

To learn the metacharacters, we'll play a simple game.  Suppose we have the following strings in a bag:

        apple
        apricot
        banana 
        pineapple

When we create a new regular expression, we'll ask which of the strings match the regular expresion.    Sounds fun, right?

### The string anchors `^` and `$`
**The '^' and '$' match the beginning and end of a string, respectively.** Note that these metacharacters match a *position* in a string rather than a character.

* If the regular expression is '`^a`', which strings in our bag will match?

     * <span style="color:red">a</span>pple
     * <span style="color:red">a</span>pricot
 
 These strings match because they all start wtih a.  The regular expression doesn't match banana and pineapple because they don't start with a.
 

* If the regular expression is '`a$`', which strings in our bag will match?

     * banan<span style="color:red">a</span>
     
We can combine both `^` and `$` in the same regular expression.  This will require the pattern to match the entire string:

* If the regular expression is '`^apple$`', which strings in our bag will match?

     * <span style="color:red">apple</span>
     
 Note that even though pine<span style="color:red">apple</span> contains <span style="color:red">apple</span>, pineapple doesn't start and end with apple so it doesn't match.


### The **character class** `[]`
**The `[]` matches any one of the characters contained in the square brackets.**

* If the regular expression is '`ap[pr]`', which strings in our bag will match?

     * <span style="color:red">app</span>le
     * <span style="color:red">apr</span>icot
     * pineap<span style="color:red">app</span>le
     
 These strings match because they all contain `ap` followed by a `p` or an `r`.

If the regular expression is `[^ ... ]` instead of `[ ... ]`, the character class will match any character that is *not* listed.  Note that when `^` is inside `[]`, it means *match any character except the ones enclosed in these `[]`*, but when `^` is at the beginning of a regular expression, it mean *match at the beginning of the string*.

* If the regular expression is '`ap[^p]`', which strings in our bag will match?

     * <span style="color:red">apr</span>icot
  
  This string matches because it contains `ap` followed by a letter that is not `p`.
  
One final note on the character class `[]`: almost all metacharacters lose their special power when inside, e.g., a `$` matches a literal `$`.

### The **character class metacharacter** `-`
**When `-` is inside a character class `[]`, it indicates a range of characters.  The range must have a beginning and an end.**

The character class metacharacter `-` is useful because it saves us a lot of typing.  Suppose we wanted to find all strings that start with an uppercase letter or a number.  Without using `-`, our regular expression would be quite long: `^[ABCDEFGHIJKLMNOPQRSTUVWXYZ0123456789]`.  If we use `-`, the regular expression is much shorter: `^[A-Z0-9].

Again, note that context matters: when `-` is inside `[]`, it means *match any character in the range*, but when `-` outside `[]`, it means *match a literal `-`*.

* If the regular expression is '`[r-u]$`', which strings in our bag will match?

     * aprico<span style="color:red">t</span>
     
  This string matches because it ends in either `r`, `s`, `t`, or `u`.
  
### The `.` character class
** The `.` matches any character.**

* If the regular expression is '`b.n`', which strings in our bag will match?

     * <span style="color:red">ban</span>ana
     
  This string matches because it contains a `b`, then any character, then `n`.
  
### Other special character classes
There are a few sets of characters that are often searched for.  We could use `[]` to represent each set, e.g., to match a digit we could use `[0-9]`.  However, there is an even shorter way to match a digit, and this is with the special character class `\d`.  Below is a table of some common character classes.

Character class | Description | Longhand version
:---: | --- | ---
`\d` | Match a digit | `[0-9]`
`\D` | Match a nondigit | `[^0-9]`
`\s` | Match a whitespace character, e.g., a tab, a space, etc. | `[ \t\n\r\f\v]`
`\S` | Match a nonwhitespace character | `[^ \t\n\r\f\v]`
`\w` | Match a **word** character.  A word character is any letter, number, or `_` | `[a-zA-Z0-9_]`
`\W` | Match a nonword character. |`[^a-zA-Z0-9_]`

  
### The alternation `|` metacharacter
** The `|` metacharacter means or.  It designates that either the regular expression to the left of `|` or the regular expression to the right can match.**

* If the regular expression is '`apricot|banana`', which strings in our bag will match?

     * <span style="color:red">apricot</span>
     * <span style="color:red">banana</span>
     
  These strings match because they contain `apricot` or `banana`.
  
### The quantifiers
#### Match optional items with `?`
** The `?` metacharacter means optional.  It designates that the previous item is allowed, but not required.**

* If the regular expression is '`apricots?`', which strings in our bag will match?

     * <span style="color:red">apricot</span>

  This string matches because it contains `apricot` followed by an optional `s` character.
  
#### Match 0 or more items with `*`
** The `*` metacharacter matches 0 or more of the previous item.**

* If the regular expression is '`.*`', which strings in our bag will match?

     * <span style="color:red">apple</span>
     * <span style="color:red">apricot</span>
     * <span style="color:red">banana</span>
     * <span style="color:red">pineapple</span>

  These strings match because they all contain 0 or more of any character.  **.\*** is a common regular expression because it matches any string, even empty ones!
  
#### Match 1 or more items with `+`
** The `+` metacharacter matches 1 or more of the previous item.**

* If the regular expression is '`p?`', which strings in our bag will match?

     * a<span style="color:red">pp</span>le
     * a<span style="color:red">p</span>ricot
     * <span style="color:red">p</span>inea<span style="color:red">pp</span>le

  These strings match because they all contain 1 or more `p`'s.
  

#### Match between n and m items with `{}`
** The `{n,m}` metasequence matches at minimum n to at most most m of the previous item.**

* If the regular expression is '`p{3,5}`', which strings in our bag will match?

  No strings match because none of the strings contain `ppp`, `pppp`, or `ppppp`.


### The escape character `\`
** The `\` metachacter is used to "escape" or ignore a metacharacter's function and instead use the literal chacter instead.**

Suppose our strings are `Mr. Happy` and `Mrs. Glad`.  The regular expression '`Mr.`' would match

   * <span style="color:red">Mr.</span> Happy
   * <span style="color:red">Mrs</span>. Glad

because these strings contain `Mr` and then any character.  

If we escape the `.` and use the regular expression '`Mr\.`', it would match
  
   * <span style="color:red">Mr.</span> Happy

because the string contains `Mr.` `Mrs. Glad` no longer matches.


### Capturing matches with `()`
Oftentimes, you want to return the substring that matched a specific part of a regular expression.  **The `()` metasequence captures a particular subset of a regular expression**  This process is also called grouping.

Suppose again we have the `Mr. Happy` and `Mrs. Glad` strings.  We want to write a  regular expression that will capture the surname.  The regular expression '`Mrs?\. (.*)`' will do just that.  This regular expression would match

   * <span style="color:red">Mr. </span><span style="background-color: #A0F000; color:red">Happy</span>
   * <span style="color:red">Mrs. </span><span style="background-color: #A0F000; color:red">Glad</span>

The strings match because they contain `Mr`, followed by an optional `s`, a `.`, a space, and then 0 or more of any character.  Since we enclosed the `.*` in `()`, we captured the contents of what matched that part of the regular expression.  (This was denoted with the <span style="background-color: #A0F000; color:red">green background and red text</span>.)

**Note:** Regular expressions are **greedy**.  This means they try to make the longest match possible.  Notice that in the last example, the `.*` matched all the way to the end of the string!  If our string was `Mr. Happy, so nice to see you!`, our regular expression would have matched <span style="color:red">Mr. </span><span style="background-color: #A0F000; color:red">Happy, so nice to see you!</span>.  This wouldn't have captured just the surname.

## Using regular expressions in Python
That was a long description of what regular expressions are; but hopefully you've come to realize regular expressions are extemely powerful for matching text.  Knowing how to match specific text in files is an important skill in bioinformatics.

Let's now learn how to use regular expressions in Python.

In order to use a regular expression in Python, we need to import the `re` module:

        >>> import re
        
There are two main functions when dealing with regular expressions: **`re.search`** and **`re.match`**.

The syntax for `re.search` is:

        re.search(regular_expression, string, [flags])
        
where `regular_expression` is the regular expression to match, `string` is the string you want to match with the regular expression, and `flags` are optional modifiers to change how the matching is done.  For example, the `re.I` flag performs a case-INsensitive search.

**Note:** `re.search` only searches for the *first* occurrence of the pattern within the string.  It is blind to any other occurrences within the string.  (To find all of the occurrenecs, use the [`re.findall`](https://docs.python.org/3.4/library/re.html#re.findall) function.)

Since `re.search` is a function, it returns something.  If the string matches the regular expression, it returns a [**match object**](https://docs.python.org/3/library/re.html#match-objects); if the string does not match, it returns `None`.

Here's an example:

        >>> match_obj = re.search(r'apple', "pineapple")
        >>> if match_obj:
        ...     print("The string matched")
        ... else:
        ...     print("The string did NOT match")
        ...

Notice how the regular expression is surrounded in `r''`.  This is Python's syntax for using a regular expression.

Here's an example with a metacharacter:

        >>> match_obj = re.search(r'^apple', "pineapple")
        >>> if match_obj:
        ...     print("The string matched")
        ... else:
        ...     print("The string did NOT match")
        ...
        

`re.match` is basically the same as `re.search`, except that it matches a regular expression *at the beginning of a string*.  `re.search` will search for a regular expression anywhere in a string.


There are special methods we can apply on a match object to learn more about what was matched:

        >>> match_obj = re.search(r'apple', "pineapple")
        >>> match_obj.group()
        'apple'
        >>> match_obj.span()
        (4, 9)

* **`re.group`** returns the substring matched by the regular expression
* **`re.span`** returns a tuple containing the start and end positions of the match.

We can use `()` metacharacters to capture part of a match:

        >>> match_obj = re.search(r'Mr. (.*)', "Mr. Happy")
        >>> match_obj.group(0)
        'Mr. Happy'
        >>> match_obj.group(1)
        'Happy'

Captured groups are numbered left to right and start with 0.  The first group, 0, is always the string that matched the entire regular expression.  The next group, i.e., the leftmost `()`, returns the surname, `Happy`.  There are other captured groups in the regular expression.
