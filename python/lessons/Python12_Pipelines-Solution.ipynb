  Follow the directions inside the script to learn how the pipeline works and write the code.  You will add code wherever there is a `TODO` comment.

  You'll need the `proteins.faa` file from before.  Copy it to your current working directory or download it again:
  
      $ wget http://genomic.wustl.edu/bioinformatics/proteins.faa
      
**Hints:**
* Use the Python function **`subprocess.call`** to call an external command from within a Python script.  The syntax is:

        subprocess.call(list_of_arguments)
    
  where `list_of_arguments` is a Python list of the words in the command.
    For example, if you wanted to execute
    
        $ cd-hit -i proteins.faa -o clustered_proteins.faa
        
    `list_of_arguments` would be
    
        ["cd-hit", "-i", "proteins.faa", "-o", "clustered_proteins.faa"]
        
    and the `subprocess.call` code would be
    
        subprocess.call(["cd-hit", "-i", "proteins.faa", "-o", "clustered_proteins.faa"])
        
* When you want to capture the stdout of an external command, the syntax for `subprocess.call` is a little different:

        subprocess.call(list_of_arguments, stdout=file_object)

  where `file_object` is a Python file object for a file that you opened for writing.  Refer to the previous [reading and writing files lesson](http://nbviewer.ipython.org/github/mollygibson/bioinformatics_wkshp_washu/blob/master/python/lessons/Python7_FileIO.ipynb) for help with file objects.
