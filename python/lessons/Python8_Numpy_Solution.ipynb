The most common parametric test to test if the mean gene expression is different between two independent groups is the **T-test**. We can perform this using the **ttest_ind** function from the SciPy stats module. This test "Calculates the T-test for the means of TWO INDEPENDENT samples of scores." You can look at the official documentation for this function [here](http://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.ttest_ind.html).

# Solution

1. Import all necessary scientific python packages:

        >>> import numpy as np
        >>> import scipy as sp
        >>> from scipy import stats

2. Load the control and treatment data into numpy arrays:

        >>> control = np.loadtxt("control.csv", delimiter=",")
        >>> treatment = np.loadtxt("treatment.csv", delimiter=",")
        
3. Test each group from normality:

        >>> stats.normaltest(control)
        (1.5303051917002202, 0.4652629283650207)
        >>> stats.normaltest(treatment)
        (0.12639399259220102, 0.93875852352041589)
        
    For both groups, P>0.05, so we fail to reject the null hypothesis that the sample comes from a normal distribution.
    
4. **BONUS:** Test for equal variances. Since we know each sample comes from a normal distribution, we can use [Bartlett's test](http://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.bartlett.html#scipy.stats.bartlett):

        >>> stats.bartlett(control,treatment)
        (66.036171191657232, 4.4272319673270792e-16)
        
    P < 0.05, so we reject the null hypothesis that the control and treatment samples have equal variances. Let's remember this when we get ready to test for equal mean expression.
    
    
5. Test for difference in mean expression using [**ttest_ind**](http://docs.scipy.org/doc/scipy-0.14.0/reference/generated/scipy.stats.ttest_ind.html):

        >>> stats.ttest_ind(control,treatment)
        (-35.70809652966031, 1.1290024730419147e-90)
    
    P < 0.05, so we reject the null hypothesis that the control and treatment groups have equal mean expressions.
