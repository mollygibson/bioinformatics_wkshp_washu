# Solution

## Part 1

**Remember the data we are working with and the figure as it stands:**
        >>> x = np.linspace(0, 2*np.pi)
        >>> y1 = np.sin(x)
        >>> y2 = np.cos(x)
        
        >>> plt.subplots(1, 2)
        
        >>> plt.subplot(1,2,1)
        >>> plt.plot(x, y1)
        
        >>> plt.subplot(1,2,2)
        >>> plt.plot(x, y2)

        
1. Change the x axis on ax1 to run from 0 to 4:
    
        >>> plt.subplot(1,2,1)
        >>> plt.xlim(0,4)
        
        >>> plt.subplot(1,2,2)
        >>> plt.xlim(0,4)
        
1. Add labels to the x axis on both subplots:

        >>> plt.subplot(1,2,1)
        >>> plt.xlabel("This is the first plot")
        
        >>> plt.subplot(1,2,2)
        >>> plt.xlabel("This is the second plot")
  
  
**Finally, let's save it again:**
        >>> plt.savefig('fig_with_two_plots-part1')

## Part 2

  
**This is exactly like before with one major difference:**

Make the suplots:
        >>> plt.subplots(1, 2)
        
Plot the first plot:
        >>> plt.subplot(1,2,1) 
        >>> plt.plot(x, y1, color='r', linestyle='--') ## This is the key!!
        >>> plt.xlim(0,4)
        >>> plt.xlabel("This is the first plot")
        
Plot the second plot:
        >>> plt.subplot(1,2,2)
        >>> plt.plot(x, y2)
        >>> plt.xlim(0,4)
        >>> plt.xlabel("This is the second plot")

Save the figure:
        >>> plt.savefig('fig_with_two_plots-part2')
