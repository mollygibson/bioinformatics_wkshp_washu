# What is the point?

Learning to program takes time. Why take that time away from research/lab to learn to program?

From [Why Biologist Want to Program Computers](http://archive.oreilly.com/pub/a/oreilly//news/perlbio_1001.html), there are also scientific reasons for learning how to program:

1. **Quantity of existing data: **

    There is now a huge amount of basic biological data, and without computers to store and search this data, we'd be severely constrained. 


2. **Dealing with new data: **

    Many biologists find that their laboratory notebooks still work just fine for recording their results. But for an increasing number of researchers, new laboratory techniques (high-throughput sequencing, microarrays and so on) are generating a volume of data that requires computer technology.
    

3. **Automating the automation: **

    One of the most valuable, time-saving programming skills that a biologist can learn is how to automate the automation. This means writing programs that run, and collect the output from, other programs, thereby eliminating the need to run the other programs in person while sitting at the computer. For example:

    4. **Evaluating many targets: **
    Say you have two hundred candidate targets for some biological experiment. The experiment is lengthy and costly, and so you need to evaluate the targets in order to select the most promising ones. You have several programs whose combined results allow you to make a reasonable selection. If running the needed programs, and examining and prioritizing the results, takes an hour per target, **then you've got about a month's worth of work sitting in front of a computer screen ahead of you.**

    *So instead, you write a program:* This program takes each target one at a time, then runs the auxiliary programs and collects and collates the results. After it's done, it prioritizes the targets according to the criteria you have programmed, and then it presents you with a "top ten" list of targets. The program may take a couple of days, or a week, to write; it takes 30 minutes to run. You've saved yourself the rest of the month for doing "real" biology, actually performing the experiments on the targets; and you're on track to get to the finish line three weeks sooner.

# Be Patient with Yourself
While we believe that every graduate biology student **can** and should learn at least basic computational skills, ***programming is hard***. 

Everyone should repeat with me: **programming is hard**. From ["Don't believe anyone who tells you learning to program is easy"](http://techcrunch.com/2014/05/24/dont-believe-anyone-who-tells-you-learning-to-code-is-easy/), don't let your brain do this:

<img src="programming_is_easy.jpg" width="600">

Most of us are deeply disturbed at the prospect of being horrible at something, even temporarily. When you try something new, you’re usually very bad, and you know it. So, our best advice to getting through this workshop: 

1. be ultra patient with yourself
2. be persistent
3. **remember, we are here to help!**

The early hours of learning to program are always challenging, but a little patience and persistance can result in huge increases.
