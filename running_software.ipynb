# Running bioinformatics software

So far, we have taught you a lot about how to analyze biological data by writing your own scripts. However, there is a lot of bioinformatics software out there already and one of them might help you accomplish your goal without reinventing the wheel. Often though, it is a little overwhelming to learn how to use these scripts or software. In this lesson, we will learn how to download, setup, and run a script that clusters nucelotide sequences at a particular percent identity. 

## CD-HIT

### 1. Download CD-HIT

CD-HIT source code is stored in [Google Code](https://code.google.com/p/cdhit/downloads/list). First thing we want to do is download the source code and extract the code:

1. `$ wget https://cdhit.googlecode.com/files/cd-hit-v4.6.1-2012-08-27.tgz`
    * Note: You can right click and *copy link address* in order to paste into the command line for `wget` command
    
2. `$ tar -zxvf cd-hit-v4.6.1-2012-08-27.tgz`
    * **-z** : Work on gzip compression automatically when reading archives
    * **-x** : Extract archives.
    * **-v** : Produce verbose output i.e. display progress and extracted file list on screen
    * **-f** : Read the archive from the archive to the specified file. 
    
This should result in a folder called `cd-hit-v4.6.1-2012-08-27` in the same directory.


### 3. Complie cd-hit

We have been programming in Python today, which is an [interpreted language](http://en.wikipedia.org/wiki/Interpreted_language). This is important, because we don't ever have to compile our code into "machine code" before running our scripts. A lot of software is written in [compiled languages](http://en.wikipedia.org/wiki/Compiled_language), however, because programs compiled into machine code at compile time tend to be quicker than those translated at run time, due to the overhead of the translation process.

CD-HIT is written in a compiled language, so we need to compile it before we can actually run it. In order to do this, the authors have written a [Makefile](http://en.wikipedia.org/wiki/Makefile). This is just was to build a piece of software in a predefined manner with one command. We will use their Makefile:

1. Change into the cd-hit folder you just unzipped:
        $ cd cd-hit-v4.6.1-2012-08-27
        
2. Compile the software: 
        $ make

### 2. Format your .bashrc

Before we can directly run cd-hit, we need to make sure the computer knows where you put it. To do this, we will modify your **PATH environment variable**. The PATH environment variable is a colon-delimited list of directories that your shell searches through when you enter a command.

Let's look at what your PATH variable looks like right now.

    $ echo $PATH
    
Right now, the paths in stored in this variable are the only directories that the shell will look in for the software you are attempting to run from the command line. What we want to do is add the folder we just downloaded to your PATH environment variable so we can run it directly. The PATH variable is defined in a file called the **.bashrc** file that is stored in your home directory. Let's edit this folder:

1. Change into the folder you just downloaded `cd cd-hit-v4.6.1-2012-08-27`
1. Determine the path of this folder using `pwd` and copy it.
1. Change to your home directory: `cd ~`
2. Open your `.bashrc` file using nano: `nano .bashrc`
3. Append the copied path to your PATH environment variable by adding the following line to this folder:
        PATH=<the path you copied>/cd-hit-v4.6.1-2012-08-27/:$PATH
   
    It is very important that you remember the :$PATH!

4. After you build the PATH environment variable in your `.bashrc`, you need to make sure you export it:
        export PATH

5. Finally, whenever you make a modification to your `.bashrc` file, you must "source" in order to load it or take effect. (Note: The `.bashrc` file loads everytime you log onto the machine).
        $ source .bashrc

6. Make sure it works!
        $ cd-hit --help

### 3. Run cd-hit

*Finally*, we are ready to run cd-hit! We will run cd-hit to cluster a set of protein sequences:

    wget https://cgs.wustl.edu/~mgibson/bioinformatics/proteins.faa
    
Let's first run cd-hit with all the default parameters:

    cd-hit -i proteins.faa -o clustered_proteins.faa
